```python
city_cost=[]
for city in df_j:
    cost=[]
#     print(df_j[city]['restaurants'])
    if 'restaurants' in df_j[city].keys():
        try:
            for i in df_j[city]['restaurants'].keys():
                cost.append(int(df_j[city]['restaurants'][i]['cost'][2:]))
#                 print(sum(cost))
                avg_cost=round(sum(cost)/len(cost),2)
        except:
            pass
    city_cost.append([city,avg_cost])
    
```

```python
print((city_cost))
```

```python
df_city=pd.DataFrame(city_cost,columns=['city','cost'])
```

```python
# Top 5 Expensive cities in the datset
df_city.sort_values(by='cost',ascending=False).head()
```

```python
restaurants_listed=[]
for city in df_j:
#     print(df_j[city]['restaurants'])
    x=0
#     print(x)
    try:
#         print(df_j[city]['restaurants'])
           x=len(df_j[city]['restaurants'].keys())
#         print(x)
    except:
#         raise e
        pass
    restaurants_listed.append([city,x])
    
```

```python
df_restaurants_listed=pd.DataFrame(restaurants_listed,columns=['city','Number Of Restaurants Listed'])
```

```python
most_popular={}
for city in df_j:
    if 'restaurants' in df_j[city].keys():
        try:
            for i in df_j[city]['restaurants'].keys():
                if df_j[city]['restaurants'][i]['name'] not in most_popular:
                    most_popular[df_j[city]['restaurants'][i]['name'] ]=1
                else:
                    most_popular[df_j[city]['restaurants'][i]['name'] ]+=1
        except:
            pass
```

```python
df_most_popular=pd.DataFrame()
df_most_popular['restaurants']=most_popular.keys()
df_most_popular['Number of Restaurants']=most_popular.values()
```

```python
# Most Popular Food Chains in India
df_most_popular.sort_values(by='Number of Restaurants',ascending=False).head(10)
```

```python
df_j['Bangalore'].keys()
```

```python
most_popular_pune={}
for sub in df_j['Pune'].keys():
    try:
         for i in df_j['Pune'][sub]['restaurants'].keys():
            if df_j['Pune'][sub]['restaurants'][i]['name'] not in most_popular_pune:
                most_popular_pune[df_j['Pune'][sub]['restaurants'][i]['name']]=1
            else:
                most_popular_pune[df_j['Pune'][sub]['restaurants'][i]['name']]+=1
    except :
        pass
            
```

```python
df_pune=pd.DataFrame()
df_pune['Restaurants']=most_popular_pune.keys()
df_pune['Popularity']=most_popular_pune.values()
```

```python
# Most Popular Restaurants in Pune
df_pune.sort_values(by='Popularity',ascending=False).head(10)
```

```python
most_cheap_delhi=[]
for sub in df_j['Delhi'].keys():
    cost=[]
    try:
         for i in df_j['Delhi'][sub]['restaurants'].keys():
#                 print(df_j['Delhi'][sub]['restaurants'][i]['cost'][2:])
                cost.append(int(df_j['Delhi'][sub]['restaurants'][i]['cost'][2:]))
    except :
        pass
    if not len(cost)==0:
#         print(sum(cost))
        avg_cost=sum(cost)/len(cost)
        most_cheap_delhi.append([sub,avg_cost])
            
```

```python
df_delhi_cheap=pd.DataFrame(most_cheap_delhi,columns=['SubRegion','Cost'])
```

```python
# Cheapest Restaurant in The Subregion of Delhi
df_delhi_cheap.sort_values(by='Cost',ascending=True).head(1)
```

```python
most_rated_bangalore={}
count={}
for sub in df_j['Bangalore'].keys():
    rating=[]
    try:
         for i in df_j['Bangalore'][sub]['restaurants'].keys():
                if df_j['Bangalore'][sub]['restaurants'][i]['name'] not in most_rated_bangalore:
                    most_rated_bangalore[df_j['Bangalore'][sub]['restaurants'][i]['name']]=float(df_j['Bangalore'][sub]['restaurants'][i]['rating'])
                else:
                    most_rated_bangalore[df_j['Bangalore'][sub]['restaurants'][i]['name']]+=float(df_j['Bangalore'][sub]['restaurants'][i]['rating'])
                if df_j['Bangalore'][sub]['restaurants'][i]['name'] not in count:
                    count[df_j['Bangalore'][sub]['restaurants'][i]['name']]=1
                else:
                    count[df_j['Bangalore'][sub]['restaurants'][i]['name']]+=1
    except :
        pass
            
```

```python
most_rated_bangalore
```

```python
count
```

```python
for key,value in most_rated_bangalore.items():
    value=round(value/count[key],2)
    most_rated_bangalore.update({key:value})
```

```python
most_rated_bangalore
```

```python
df_top_b=pd.DataFrame()
df_top_b['restaurants']=most_rated_bangalore.keys()
df_top_b['Rating']=most_rated_bangalore.values()
```

```python
# Top Rated Restaurants in Bangalore
df_top_b.sort_values(by='Rating',ascending=False).head(10)
```

```python
most_rated_patna={}
count_patna={}
for i in df_j['Patna']['restaurants'].keys():
    rating=[]
    try:
            if df_j['Patna']['restaurants'][i]['name'] not in most_rated_patna:
                most_rated_patna[df_j['Patna']['restaurants'][i]['name']]=float(df_j['Patna']['restaurants'][i]['rating'])
            else:
                most_rated_patna[df_j['Patna']['restaurants'][i]['name']]+=float(df_j['Patna']['restaurants'][i]['rating'])
            if df_j['Patna']['restaurants'][i]['name'] not in count_patna:
                count_patna[df_j['Patna']['restaurants'][i]['name']]=1
            else:
                count_patna[df_j['Patna']['restaurants'][i]['name']]+=1
    except :
        pass
            
```

```python
len(most_rated_patna)
```

```python
len(count_patna)
```

```python
for key,value in most_rated_patna.items():
    value=round(value/count_patna[key],2)
    most_rated_patna.update({key:value})
```

```python
df_top_patna=pd.DataFrame()
df_top_patna['restaurants']=most_rated_patna.keys()
df_top_patna['Rating']=most_rated_patna.values()
```

```python
# Top Rated Restaurants In Patna
df_top_patna.sort_values(by='Rating',ascending=False).head(10)
```
